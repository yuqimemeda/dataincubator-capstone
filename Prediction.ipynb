# Webscraping

```python
estimate=list()
pre=list()
day_delta=list()
weekday_list=list()
temp=list()
for i in range(5):
# webscrap future five days' temperature from the Internet
    url = "http://www.accuweather.com/en/us/new-york-ny/10007/daily-weather-forecast/349727?day=" + str(i+1)    
    url_response=ul.urlopen(url)
    soup= BeautifulSoup(url_response)
    if i==0:
        for table in soup.find_all('span'): #pick each link in turn
            h_table=table.get('class') #Get the class property of the link
            if h_table and 'temp' in h_table: #See if the word 'temp' is in the link
                sector_table=table
                break
        tomorrow_temp_str=sector_table.get_text()[:2]
    else:
        tomorrow_temp_str=((soup.find_all('strong'))[i+2]).get_text()[:2]
    tomorrow_temp=int(tomorrow_temp_str)    
    temp.append(tomorrow_temp)
    div1=soup.find_all('div')
    j =0
    for table in soup.find_all('div'): #pick each link in turn
        h_table=table.get('class') #Get the class property of the link
        if h_table and 'content' in h_table: #See if the word 'content' is in the link
            sector_table=table
            j=j+1
            if j == 2:
                break
    for table in sector_table.find_all('ul'): #pick each link in turn
        h_table=table.get('class') #Get the href property of the link
        if h_table and 'stats' in h_table: #See if the word 'stats' is in the link
            stats_table=table
    precipitation=(stats_table.find_all('li'))[2]
    tomorrow_pre_str=((precipitation.find_all('strong'))[0]).get_text()
    tomorrow_pre=float(tomorrow_pre_str.strip(' in'))  
    pre.append(tomorrow_pre)
    import datetime
    d = datetime.datetime.now() + datetime.timedelta(days=i)
    month1=d.month # get the month of the date
    from datetime import datetime,date
    a = datetime(2015, 1, 1,0,0)
    delta=d-a
    delta_day=delta.days
    day_delta.append(delta_day) # get the days passed since 01/01/2015
    
    if d.isoweekday() in range(1,6):
        weekday2=True
    else: 
        weekday2=False
    weekday_list.append(weekday2)
```
